```python
from feos_gc_pcsaft.si import *
from feos_gc_pcsaft.dft import *
from feos_gc_pcsaft import *

import matplotlib.pyplot as plt
```

```python
func = GcPcSaftFunctional(GcPcSaftFunctionalParameters.from_json_segments(['1-pentanol'], '../parameters/gc_substances.json', '../parameters/sauer2014_hetero.json'))
vle = PhaseEquilibrium.pure(func, 300*KELVIN)
profile = PlanarInterface.from_tanh(vle, 4096, 100*ANGSTROM, 600*KELVIN).solve()
```

```python
f, ax = plt.subplots()

ax.plot(profile.z/(NANO*METER), (profile.density*NAV*(NANO*METER)**3).T)
ax.legend(['CH$_3$'] + ['CH$_2$']*4 + ['OH'])
ax.set_xlabel('$z~~/~~\mathrm{nm}$')
ax.set_ylabel('$\\rho~~/~~\mathrm{nm}^{-3}$')
ax.axis([2.5,6,0,11])
```

```python
bulk = State(func, 300*KELVIN, pressure=5*BAR)
solver = DFTSolver().picard_iteration(tol=1e-5, beta=0.05).anderson_mixing()
pore = Pore1D(Geometry.Cartesian, 20*ANGSTROM, ExternalPotential.LJ93(3.0, 10.0, 0.08)).initialize(bulk).solve(solver)
```

```python
f, ax = plt.subplots()

ax.plot(pore.z/(NANO*METER), (pore.density*NAV*(NANO*METER)**3).T)
ax.legend(['CH$_3$'] + ['CH$_2$']*4 + ['OH'])
ax.set_xlabel('$z~~/~~\mathrm{nm}$')
ax.set_ylabel('$\\rho~~/~~\mathrm{nm}^{-3}$')
ax.axis([0,1.2,0,11])
```
